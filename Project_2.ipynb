Γράψτε στο παρακάτω code cell όλο τον Bash Script κώδικα του Aggregative Movie Preference Analyzer

```python
%%file aggregative_movie_preference_analyzer.sh

#!/bin/bash
RATINGS_FILE="ratings.dat"
MOVIES_FILE="movies.dat"

function printMenu()
{
    echo -e "=======================================================================================================\n\
                            Hi, welcome to Aggregative Movie Preference Analyzer :)\n\
=======================================================================================================\n\
Available commands:\n\
1. rating T1 T2      - Finds all the movies with mean rating greater than T1 and less or equal than T2.\n\
2. top_movies K      - Finds K movies with the highest mean rating.\n\
3. dominance         - Find movies that are not dominated by others.\n\
4. iceberg K T       - Find movies with at least K reviews and an average rating above T.\n\
5. top_user K        - Find the K users who have rated the most movies.\n\
6. exit or e         - Exit the program.\n\
======================================================================================================="
}


function rating() {

    T1="$1"
    T2="$2"

    declare -A movieRatings
    declare -A movieCount
    declare -A movieTitles

    while IFS=":" read -r user_id _ movie_id _ rating _ _
    do
      if [ -z "${movieRatings[$movie_id]}" ]
      then
        movieRatings["$movie_id"]="$rating"
        movieCount["$movie_id"]="1"
      else
        movieRatings["$movie_id"]=$(echo "${movieRatings[$movie_id]} + $rating" | bc)
        movieCount["$movie_id"]=$((movieCount["$movie_id"] + 1))
      fi

    done < "$RATINGS_FILE"

    while IFS=":" read -r movie_id _ title _ genre
    do
      movieTitles["$movie_id"]="$title"

    done < "$MOVIES_FILE"

    echo "-------------------------------------------------------"
    echo "Movie's Title                            Average Rating"
    echo "-------------------------------------------------------"

    found=0
    for i in "${!movieRatings[@]}"
    do
        mean=$(echo "${movieRatings[$i]} / ${movieCount[$i]}" | bc -l)
        if (( $(echo "$mean > $T1" | bc -l) )) && (( $(echo "$mean <= $T2" | bc -l) ))
        then
            title="${movieTitles[$i]}"
            printf "%-40s %.2f\n" "$title" "$mean"
            found=1
        fi
    done

    if [ "$found" -eq 0 ]
    then
        echo -e "There are no movies with average rating between $T1 and $T2."
    fi
}

function top_movies()
{
    K="$1"

    declare -A movieRatings
    declare -A movieCount
    declare -A movieMeans
    declare -A movieTitles

    while IFS=":" read -r user_id _ movie_id _ rating _ _
    do
      if [ -z "${movieRatings[$movie_id]}" ]
      then
        movieRatings["$movie_id"]="$rating"
        movieCount["$movie_id"]="1"
      else
        movieRatings["$movie_id"]=$(echo "${movieRatings[$movie_id]} + $rating" | bc)
        movieCount["$movie_id"]=$((movieCount["$movie_id"] + 1))
      fi

    done < "$RATINGS_FILE"

    while IFS=":" read -r movie_id _ title _ genre
    do
      movieTitles["$movie_id"]="$title"

    done < "$MOVIES_FILE"

    for i in "${!movieRatings[@]}"
    do
        mean=$(echo "${movieRatings[$i]} / ${movieCount[$i]}" | bc -l)
        movieMeans["$i"]="$mean"
    done


    echo "-------------------------------------------------------"
    echo "Movie's Title                            Average Rating"
    echo "-------------------------------------------------------"

    for i in "${!movieMeans[@]}"
    do
        echo "$i ${movieMeans[$i]}"
    done | sort -k2,2nr -k1,1 | head -n "$K" | while read -r movie_id mean
    do
        title="${movieTitles[$movie_id]}"
        printf "%-40s %.2f\n" "$title" "$mean"
    done

}

function dominance()
{
    declare -A movieRatings
    declare -A movieCount
    declare -A movieMeans
    declare -A movieTitles

    while IFS=":" read -r _ _ movie_id _ rating _ _
    do
        if [ -z "${movieRatings[$movie_id]}" ]
        then
            movieRatings["$movie_id"]="$rating"
            movieCount["$movie_id"]="1"
        else
            movieRatings["$movie_id"]=$(echo "${movieRatings[$movie_id]} + $rating" | bc)
            movieCount["$movie_id"]=$((movieCount["$movie_id"] + 1))
        fi
    done < "$RATINGS_FILE"

    while IFS=":" read -r movie_id _ title _ genre
    do
      movieTitles["$movie_id"]="$title"

    done < "$MOVIES_FILE"

    for i in "${!movieRatings[@]}"
    do
        movieMeans["$i"]=$(echo "${movieRatings[$i]} / ${movieCount[$i]}" | bc -l)
    done

    declare -A dominated
    for M1 in "${!movieMeans[@]}"
    do
        for M2 in "${!movieMeans[@]}"
        do
            if [ "$M1" != "$M2" ]
            then
                if (( $(echo "${movieMeans[$M1]} <= ${movieMeans[$M2]}" | bc -l) )) && [ "${movieCount[$M1]}" -le "${movieCount[$M2]}" ]
                then
                    dominated["$M1"]=1
                    break
                fi
            fi
        done
    done

    echo "-------------------------------------------------------------------------------------------------"
    printf "%-50s %-20s %-10s\n" "Movie's Title" "Average Rating" "Num Of Ratings"
    echo "-------------------------------------------------------------------------------------------------"

    for M1 in "${!movieMeans[@]}"
    do
        if [ -z "${dominated[$M1]}" ]
        then
            mean=$(echo "scale=2; ${movieRatings[$M1]} / ${movieCount[$M1]}" | bc)
            printf "%-50s %-20.2f %-10d\n" "${movieTitles[$M1]}" "$mean" "${movieCount[$M1]}"
        fi
    done
}

function iceberg()
{
    K="$1"
    T="$2"

    declare -A movieRatings
    declare -A movieCount

    while IFS=":" read -r user_id _ movie_id _ rating _ _
    do
      if [ -z "${movieRatings[$movie_id]}" ]
      then
        movieRatings["$movie_id"]="$rating"
        movieCount["$movie_id"]="1"
      else
        movieRatings["$movie_id"]=$(echo "${movieRatings[$movie_id]} + $rating" | bc)
        movieCount["$movie_id"]=$((movieCount["$movie_id"] + 1))
      fi

    done < "$RATINGS_FILE"

    while IFS=":" read -r movie_id _ title _ genre
    do
      movieTitles["$movie_id"]="$title"

    done < "$MOVIES_FILE"

    echo "-------------------------------------------------------------------------------------------------"
    printf "%-50s %-20s %-10s\n" "Movie's Title" "Average Rating" "Num Of Ratings"
    echo "-------------------------------------------------------------------------------------------------"

    found=0
    for i in "${!movieRatings[@]}"
    do
        mean=$(echo "${movieRatings[$i]} / ${movieCount[$i]}" | bc -l)
        num_of_ratings="${movieCount[$i]}"
        if (( $(echo "$mean > $T" | bc -l) )) && (( $(echo "$num_of_ratings >= $K" | bc -l) ))
        then
            printf "%-50s %-20.2f %-10d\n" "${movieTitles[$i]}" "$mean" "$num_of_ratings"
            found=1
        fi
    done

    if [ "$found" -eq 0 ]
    then
        echo -e "There are no movies with average rating over $T and Num Of Ratings equal or more $K."
    fi

}

function top_user()
{
    K="$1"

    declare -A user_ratings


    while IFS=":" read -r user_id _ movie_id _ rating _ _
    do
      if [ -z "${user_ratings[$user_id]}" ]
      then
        user_ratings["$user_id"]="1"
      else
        user_ratings["$user_id"]=$((user_ratings["$user_id"] + 1))
      fi

    done < "$RATINGS_FILE"

    echo "------------------------------------------------------"
    printf "%-40s %-40s\n" "UserID" "Num Of Movies"
    echo "------------------------------------------------------"

    for i in "${!user_ratings[@]}"
    do
        echo "$i ${user_ratings[$i]}"
    done | sort -k2,2nr -k1,1 | head -n "$K" | while read -r user_id num
    do
        printf "%-40s %-40s\n" "$user_id" "$num"
    done
}

function exit()
{
    echo -e "Thank you for using our program !!! ok bye :)"

}


function main() {
    while true; do
        printMenu
        echo -n "Type your command: "
        read command arg1 arg2
        case $command in
        rating)
          rating "$arg1" "$arg2"
          ;;
        top_movies)
          top_movies "$arg1"
          ;;
        dominance)
          dominance
          ;;
        iceberg)
          iceberg "$arg1" "$arg2"
          ;;
        top_user)
          top_user "$arg1"
          ;;
        exit|e)
          exit
          break
          ;;
        *)
          echo "Invalid Command"
          ;;
        esac
    done
}

main

```

## ⏩ Bash Script Κώδικας του Movies Sentiment Analyser

```python
%%file movies_sentiment_analyser.sh
#!/bin/bash
FILE_TAGS="tags_stemmed_tokenized_clean_no_stopwords.dat"
FILE_POSITIVE="positive_stems.txt"
FILE_NEGATIVE="negative_stems.txt"

declare -A positive_counts
declare -A negative_counts
declare -A movie_ids

declare -A positive_tags
declare -A negative_tags

while read -r line
do
    positive_tags["$line"]=1
done < "$FILE_POSITIVE"

while read -r line
do
    negative_tags["$line"]=1
done < "$FILE_NEGATIVE"



while IFS=":" read -r user_id _ movie_id _ tags _ _;
do
    if [ -z "${movie_ids["$movie_id"]}" ]; then
        movie_ids["$movie_id"]=1
    fi

    # read all the tags in the tag field
    IFS=' ' read -ra tag_array <<< "$tags"

    for tag in "${tag_array[@]}"; do
        if [[ -v positive_tags["$tag"] ]]; then
            positive_counts["$movie_id"]=$((positive_counts["$movie_id"] + 1))
        fi

        if [[ -v negative_tags["$tag"] ]]; then
            negative_counts["$movie_id"]=$((negative_counts["$movie_id"] + 1))
        fi
    done
done < "$FILE_TAGS"

echo -e "movie_id\tpositive_count\tnegative_count\tsentiment"

for movie_id in "${!movie_ids[@]}"
do
    if [ -z "${positive_counts["$movie_id"]}" ]; then
        positive_counts["$movie_id"]=0
    fi

    if [ -z "${negative_counts["$movie_id"]}" ]; then
        negative_counts["$movie_id"]=0
    fi

    # Positive Tags > Negative Tags
    if [ "${positive_counts["$movie_id"]}" -gt "${negative_counts["$movie_id"]}" ]
    then
        sentiment="positive"
    # Negative Tags > Positive Tags
    elif [ "${negative_counts["$movie_id"]}" -gt "${positive_counts["$movie_id"]}" ]
    then
        sentiment="negative"
    # Positive Tags == Negative Tags
    elif [ "${negative_counts["$movie_id"]}" -eq "${positive_counts["$movie_id"]}" ]
    then
        sentiment="neutral"
    fi

    echo -e "$movie_id \t\t ${positive_counts["$movie_id"]} \t\t ${negative_counts["$movie_id"]}\t\t$sentiment"
done
```

## ⏩ Κώδικας Complex Tree Process Generator

Γράψτε στο παράτω code cell όλο τον κώδικα του Christmas Tree Process Generator.



```python
%%file complex_tree.py
import os,sys

def main():
    parent_id = os.getpid()
    print(f"Root Process [0] with id: {parent_id}")

    # Level 1
    child_index = fork_children(['1.1', '1.2', '1.3','1.4'], parent_id)

    # Level 2
    if child_index == '1.1':
        child_index = fork_children(['2.1', '2.2','2.3'], os.getpid())
    elif child_index == '1.2':
        child_index = fork_children(['2.4','2.5','2.6'], os.getpid())
    elif child_index == '1.3':
        child_index = fork_children(['2.7','2.8','2.9'], os.getpid())
    elif child_index == '1.4':
        child_index = fork_children(['2.10','2.11','2.12'], os.getpid())

    # Level 3
    if child_index == '2.1':
        child_index = fork_children(['3.1','3.2'], os.getpid())
    elif child_index == '2.3':
        child_index = fork_children(['3.3', '3.4'], os.getpid())
    elif child_index == '2.5':
        child_index = fork_children(['3.5', '3.6'], os.getpid())
    elif child_index == '2.7':
        child_index = fork_children(['3.7', '3.8'], os.getpid())
    elif child_index == '2.9':
        child_index = fork_children(['3.9', '3.10'], os.getpid())
    elif child_index == '2.11':
        child_index = fork_children(['3.11', '3.12'], os.getpid())

    # Level 4
    if child_index == '3.2':
        fork_children(['4.1'], os.getpid())
    elif child_index == '3.4':
        fork_children(['4.2'], os.getpid())
    elif child_index == '3.6':
        fork_children(['4.3'], os.getpid())
    elif child_index == '3.8':
        fork_children(['4.4'], os.getpid())
    elif child_index == '3.10':
        fork_children(['4.5'], os.getpid())
    elif child_index == '3.12':
        fork_children(['4.6'], os.getpid())

def fork_children(child_id_array, parent_id):

    for child_id in child_id_array:

        pid = os.fork()
        if pid == 0:
            # Child process
            print(f"Child[{child_id}] with id: {os.getpid()} and its Parent id: {parent_id}")
            return child_id
        elif pid > 0:
            # Parent process waits for the child process to finish.
            os.wait()
        else:
            print("Fork failed!")
            os._exit(1)

if __name__ == "__main__":
     sys.exit(main())
```

## ⏩ Κώδικας Mixed Piping

Γράψτε στο παρακάτω code cell όλο τον κώδικα του three-way pipe σας.

```python
%%file mixed_piping.py
import os
import sys

MESSAGES = ("I love PLH211          ", "The professor is boring", "but the curriculum is  ", "interesting & useful   ")

def source_process(fifoname):
    """Source Process: writes messages from MESSAGES to the named pipe."""
    try:
        # Open FIFO for writing (os.open is required for low-level file operations)
        pipe_out = os.open(fifoname, os.O_WRONLY)

        for message in MESSAGES:
            os.write(pipe_out, (message).encode()+ b"\n") # Write message as bytes
        os.close(pipe_out) # Close pipe when done
        os._exit(0)
    except Exception as e:
        print(f"Source process error: {e}", file=sys.stderr)
        os._exit(1)

def transformer_process(fifoname, p):
    """Transformer_process: reads messages from the named pipe, tokenizes them, and writes tokens to the write end of the anonymous pipe."""
    try:
        with open(fifoname, 'r') as pipein, os.fdopen(p[1], 'w') as pipeout:
            for line in pipein:
                tokens = line.strip().split()   # Read line and tokenizes it based on wite space
                for token in tokens:
                    os.write(p[1], token.encode() + b"\n")  # Write each token to the anonymous pipe
        os._exit(0)
    except Exception as e:
        print(f"Transformer process error: {e}", file=sys.stderr)
        os._exit(1)

def output_process(p):
    """
    Output process: reads tokens from the anonymous pipe and prints them to the screen.
    """
    try:
        with os.fdopen(p[0], 'r') as pipein:
            for token in pipein:
                print(token.strip())
        os._exit(0)
    except Exception as e:
        print(f"Output process error: {e}", file=sys.stderr)
        os._exit(1)

def main():
    fifo_name = "message_fifo"

    try:
        # FIFO
        if not os.path.exists(fifo_name):
            os.mkfifo(fifo_name)

        # Anonymous Pipe
        p = os.pipe()  #p[0] is the reader and p[1] is the writer in the returned tuple of os.pipe()

        if not p:
            print ("pipe call!")
            os._exit(1)

        # Source process
        if os.fork() == 0:
            source_process(fifo_name)

        # Transformer process
        if os.fork() == 0:
            os.close(p[0])  # Close read end of the anonymous pipe
            transformer_process(fifo_name, p)

        # Output process
        if os.fork() == 0:
            os.close(p[1])  # Close write end of the anonymous pipe
            output_process(p)

        os.close(p[0])
        os.close(p[1])

        os.wait()
        os.wait()
        os.wait()

    except Exception as e:
        print(f"Main process error: {e}", file=sys.stderr)

if __name__ == "__main__":
    sys.exit(main())
```

## ⏩ Τεκμηρίωση και Σχολιασμός Κώδικα Mixed Piping

Γράψτε στο παρακάτω text cell αναφορά που θα περιγράφετε τι κάνει ο κώδικας που αναπτύξατε για το mixed piping implementations σας.

Το πρόγραμμά μας αποτελείται από τρεις διεργασίες, οι οποίες χρησιμοποιούν ένα named pipe(FIFO) και ένα anonymous pipe.

Στο *MESSAGES* αποθηκεύονται τα στοιχεία του tuple.

Οι διεργασίες είναι οι εξής:
* **source_process:**
Λαμβάνει ως όρισμα το named pipe.
Γράφει κάθε στοιχείο του MESSAGES στο named pipe.

* **transformer_process:**
Λαμβάνει ως ορίσματα το named pipe και το anomymous pipe.
Διαβάζει κάθε στοιχείο από το named pipe, το χωρίζει σε tokens με βάση το space και το γράφει στο write end p[1] του anonymous pipe.

* **output_process:**
Λαμβάνει ως όρισμα το anonymous pipe.
Διαβάζει τα tokens από τo read end p[0] του anonymous pipe και τα εκτυπώνει.

Στην συνέχεια, η *main* είναι υπεύθυνη για την δημιουργία των pipes και την διαχείριση των διεργασιών. Μέσω του *os.mkfifo()* δημιουργεί το named pipe και μέσω της *os.pipe()* δημιουργεί το anonymous pipe, το οποίο αποτελείται από το read end(p[0]) και το write end(p[1]). Μέσω της os.fork() δημιουργεί τρεις διεργασίες. Η πρώτη διεργασία καλεί την source_process, η δεύτερη διεργασία καλεί την transformer_process και η τρίτη καλεί την output_process. Επιπλέον, καλεί την os.wait() για να διασφαλίσει ότι έχουν ολοκληρωθεί όλες οι διεργασίες.
